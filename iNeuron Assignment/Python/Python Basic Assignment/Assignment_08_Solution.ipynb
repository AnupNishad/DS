```python
#Solution 3

'''
    1. inputInt() - This function Accepts an integer value. 
                    This takes additional parameters min, max, greaterThan and lessThan for bounds & always returns an int.
                    
        e.g : 
            import pyinputplus as pyip
            NumInt = pyip.inputInt()
            print(NumInt)
            print(type(NumInt))
            
            Output : 
            ------------------
             21
            <class 'int'>

    2. inputFloat() - This function Accepts a floating-point numeric value. 
                      this also takes additional min, max, greaterThan and lessThan parameters & always returns a float.
                      
       e.g : 
            import pyinputplus as pyip
            NumFloat = pyip.inputFloat()
            print(NumFloat)
            print(type(NumFloat))
            
            Output : 
            ------------------
             21.0
            <class 'float'>
'''
```

```python
#Solution 4

'''

    We have inputInt() method under PyInputPlus module which only returns only integer values. 
    To restrict the input between 0 and 99, we would use parameters 
    like min & max to ensure that user enters the values between the defined range only.
    
    e.g : 
           import pyinputplus as pyip
           NumInt = pyip.inputInt("Enter a Number : ", min=0, max=100)
           print(NumInt)
            
            Output : 
            ------------------
             Enter a Number : 
             101
            Number must be at maximum 100.
            Enter a Number : 
             -01
            Number must be at minimum 0.
            Enter a Number : 
             50
            50
'''
```
