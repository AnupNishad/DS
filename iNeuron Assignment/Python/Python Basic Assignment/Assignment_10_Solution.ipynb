```python
#Solution 1

'''
    1. shutil.copy()      -  This function will copy a single file.
                             It takes two arguments src,dest, src represents the file to be copied & 
                             destination refers to the file/folder to where the src data should be copied
                             
    2. shutil.copytree()  -  This function will copy an entire folder along with all its contents.
                             It also takes two arguments src & dest, it copies all the content recursively and stores it in dest. 
                             Important point here is dest must not exist prior to this and it will be created during the copy operation.
                             
                             
    e.g-->1 : 
    
        import shutil, os
        shutil.copy("/home/jovyan/demo/data/shutilcopytestsrc.txt","/home/jovyan/demo/notebooks")
        
        Output : 
        ---------------------------------------------------
        '/home/jovyan/demo/notebooks/shutilcopytestsrc.txt'
        
        In above program file 'shutilcopytestsrc.txt' copied from "/home/jovyan/demo/data" folder
        to the destination folder "/home/jovyan/demo/notebooks"
        
        
    e.g-->2 : 
    
        import shutil, os
        shutil.copytree("/home/jovyan/demo/data","/home/jovyan/demo/CopyTreeCheck")
        
        Output : 
        ---------------------------------------------------
        '/home/jovyan/demo/CopyTreeCheck'
        
        In above program all content/files of folder 'data' copied
        to the destination folder "/home/jovyan/demo/CopyTreeCheck".
'''
```

```python
#Solution 5

import shutil,os

def SearchFileAndCopyToAnotherFolder(Source,Destination,FileExtension):
    ''' This function will filters the file types from specified path & file extension & copy those file to new folder. '''
    try:
        for root,dire,files in os.walk(Source):
              for filename in files:
                    if filename.endswith(FileExtension):
                        shutil.copy(Source+"/"+filename, Destination)
    except Exception as e:
        print("Error : ",e)

Extensions = (".pdf",".jpg")
SearchFileAndCopyToAnotherFolder("/home/jovyan/demo/data","/home/jovyan/demo/notebooks",Extensions)
```
