```python
#Solution 3
'''
    Truth Table - AND Operator
    --------------------------
                                                
    Boolean Value 1       | Boolean Value 1         | Result
    ---------------------------------------------------------
    TRUE                  | TRUE                    | TRUE
    ---------------------------------------------------------
    TRUE                  | FALSE                   | FALSE
   ----------------------------------------------------------
    FALSE                 | TRUE                    | FALSE
    ---------------------------------------------------------
    FALSE                 | FALSE                   | FALSE

    
    Truth Table - OR Operator
    --------------------------
    
    Boolean Value 1       | Boolean Value 1         | Result
    ---------------------------------------------------------
    TRUE                  | TRUE                    | TRUE
    ---------------------------------------------------------
    TRUE                  | FALSE                   | TRUE
   ----------------------------------------------------------
    FALSE                 | TRUE                    | TRUE
    ---------------------------------------------------------
    FALSE                 | FALSE                   | FALSE
    
    
    Truth Table - NOT Operator
    --------------------------
    
    Boolean Value         | Result
    -------------------------------
    TRUE                  | FALSE 
    -------------------------------
    FALSE                 | TRUE 
    
'''
```
