```python
#Solution 5
'''
    Group 0  - This will return the entire match.
    Group 1  - This will return the first set of parentheses. 
    Group 2  - This will return the second set of parentheses.
    
    We can understand better with below example :
    
    import re
    Char = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
    SearchChar = Char.search('This is my first python program 123-456-7890')
    print(SearchChar.group()) #This will return entire match string
    print(SearchChar.group(1)) #This will return first match string
    print(SearchChar.group(2)) #This will return second match string
    
    Output : 
    -----------
    123-456-7890
    123
    456-7890
    
'''
```

```python
#Solution 6
'''
    The \. \( and \) escape characters in the raw string passed to re.compile() will match actual parenthesis characters.
    Periods and parentheses can be escaped with a backslash: \., \(, and \).
    
    We can understand better with below example :
    
    import re
    Char = re.compile(r'(\(\d\d\d\)) (\d\d\d-\d\d\d\d)')
    SearchChar = Char.search('This is my first python program (123) 456-7890')
    print(SearchChar.group())
    
    Output : 
    -----------
    (123) 456-7890
    
'''
```

```python
#Solution 7
'''
    if the regex pattern has groups a list of tuple of strings is returned.
    If the regex pattern has no groups a list of strings matched is returned.
    
    We can understand better with below example :
    
    Program 1 : 
    -----------------
    import re
    Char = re.compile(r'(\(\d\d\d\)) (\d\d\d-\d\d\d\d)')
    SearchChar = Char.findall('This is my first python program (123) 456-7890')
    print(SearchChar)
    
    Output : 
    -----------
    [('(123)', '456-7890')]
    
    
    Program 2 : 
    -----------------
    import re
    Char = re.compile(r'\d{3}-\d{3}-\d{4}')
    SearchChar = Char.findall('This is my first python program 123-456-7890')
    print(SearchChar)
    
    Output : 
    -----------
    ['123-456-7890']
    
'''
```

```python
#Solution 9
'''
    9. In regular expressions, what does the character stand for?
    In above assignment question not able to find the character on which the question asked.
    
    if its about Question Mark (?) then my answer is as below : 
    
    The '?' character can either mean  - match zero or one of the preceding group.
    
    For Example :
    
    import re
    M1 = re.search("(Bo)?b","Bob returns")
    print(M1)
    M2 = re.search("Al(ic)?e","Alice returns")
    print(M2)
    
    Output : 
    --------------------
    <re.Match object; span=(0, 3), match='Bob'>
    <re.Match object; span=(0, 5), match='Alice'>
    
'''
```

```python
#Solution 10
'''
    1. + - This matches one or more.
    2. * - This matches zero or more.
    
    For Example : 
    
    import re
    M1 = re.search("(Bo)+b","Bob returns")
    print(M1)
    M2 = re.search("Al(ic)*e","Bob returns")
    print(M2)
    
    Output : 
    ---------------
    <re.Match object; span=(0, 3), match='Bob'>
    None
    
'''
```

```python
#Solution 11
'''
    1. {4}  - This matches exactly three instances of the preceding group. 
    2. {4,5} - This matches between three and five instances.
    
    For Example : 
    
    import re
    Char = re.compile(r'(Ok){4,5}')
    M1 = Char.search('OkOkOkOk')#M1 = Char.search('OkOkOkOkOk') - It can match 4 to 5
    M2 = Char.search('Ha')
    print(M1.group())
    print(M2)
    
    Output : 
    -----------------------
    OkOkOkOk
    None
'''
```
