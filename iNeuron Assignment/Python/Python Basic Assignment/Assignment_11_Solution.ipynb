```python
#Solution 1
'''
    We can create an assert statement that throws an AssertionError if the variable spam is a negative integer as follow : 
    
    spam = -1
    assert spam > 0 , 'The spam variable is less than 0.'
    print('The spam variable is greater than 0.')
    
    In above code will throw assertion error & will not execute print statement.
    
    AssertionError: The spam variable is less than 0.
    
'''
```

```python
#Solution 2
'''
    We can create an assert statement that triggers an AssertionError 
    if the variables eggs and bacon contain strings that are the same as each other, 
    even if their cases are different as follow : 
    
    eggs = "goodbye"
    bacon = "goodbye" OR "GOODbye"
    assert eggs.lower() != bacon.lower(), 'The eggs and bacon variables are the same' 
    print('The eggs and bacon variables are not the same')
    -----------------OR---------------------
    eggs = "goodbye"
    bacon = "goodbye" OR "GOODbye"
    assert eggs.upper() != bacon.upper(), 'The eggs and bacon variables are the same' 
    print('The eggs and bacon variables are not the same')
    
    In above code will throw assertion error & will not execute print statement.
    
    AssertionError: The eggs and bacon variables are the same
    
'''
```

```python
#Solution 7
'''
    Following line of code we will add to our software to disable all logging messages : 
    
    logging.disable(logging.CRITICAL)
    
'''
```

```python
#Solution 8
'''
    Following advantages of logging over print function :
    
    1. We can disable logging messages without removing the logging function calls. 
    2. We can selectively disable lower-level logging messages. 
    3. We can create logging messages which provides a timestamp.
    4. We can create a log file using logging which helps us to investigate issue occured in past.
    
'''
```
