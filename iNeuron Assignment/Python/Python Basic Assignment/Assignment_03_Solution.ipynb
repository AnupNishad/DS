```python
#Solution 7
'''
     If we take below example : 
     def add(a,b):
         print(a+b)
     a = add(2,8)
     Here variable 'a' will have value as 10 but NoneType
     If we try to add value of a with 10 will get error , because we are trying to add NoneType with int.
     
     But instead of 'print' if we use 'return' , will get type what we want to return as below
     def add(a,b):
          return (a+b)
     a = add(2,8)
     Here if we add value of a with 10 will get 20 as output, No Error.
     
     So if we use 'return' in our function it will return value with proper data type & that value we can use in future.
     
     Yes, possible to have a return value in an expression.
     
     
'''
```
