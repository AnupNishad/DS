```python
#Solution 1

def GetFactorial(Number):
    '''This function will take a integer input & provide factorial of that number as output
    '''
    try:
        j = 1
        for i in range(Number,0,-1):
            j = i*j
        print("Factorial for provided number is :",j)
    except Exception as e:
        print(e)
        
GetFactorial(5)
```

```python
#Solution 2
def GetMultiplicationTable(Number):
    '''This function will take a integer input & provide multiplication table  of that number as output
    '''
    try:
        j = Number
        for i in range(1,11):
            print(f'{j} x {i} = {i*j}')
    except Exception as e:
        print(e)
        
GetMultiplicationTable(7)
```

```python
#Solution 3

def GetFibonacciSeries(Number):
    '''This function will take a integer input & provide fibonacci sequence till that number as output
    '''
    try:
        f1=0
        f2=1
        print(f1,f2,end=" ")
        for x in range(2,Number+1):
            nextsequence=f1+f2                           
            print(nextsequence,end=" ")
            f1=f2
            f2=nextsequence
    except Exception as e:
        print(e)
        
GetFibonacciSeries(10)
```

```python
#Solution 4

def CheckArmstrongNumber(Number):
    '''This function will take a integer input & provide number is Armstrong or not as output
    '''
    try:
        sum = 0
        OriginalNumber = Number
        i = Number
        while i>0:
            digit = i%10
            sum = sum+(digit**3)
            i = i//10
        if sum == OriginalNumber:
            print(Number,"is a Armstrong Number")
        else:
            print(Number,"is not a Armstrong Number")
    
    except Exception as e:
        print(e)
        
CheckArmstrongNumber(153)
```

```python
#Solution 6

def GetSumOfNaturalNumber(Number):
    '''This function will take a integer input & provide sum of all natural number till input number as output
    '''
    try:
        sum = 0
        for i in range(1,Number+1):
            sum = sum+i
        print(sum,end=" ")
    except Exception as e:
        print(e)
        
GetSumOfNaturalNumber(20)
```
