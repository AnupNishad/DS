```python
#Solution 2

def GetAddition(a,b):
    '''This funtion will take 2 integer as input & provide result as Addition of those 2 numbers.
    '''
    try:
        return (a+b)
    except Exception as e:
        print(e)
GetAddition(2,5)

def GetDivision(a,b):
    '''This funtion will take 2 integer as input & provide result as Division of those 2 numbers.
    '''
    try:
        return (a/b)
    except Exception as e:
        print(e)
GetDivision(2,5)
```

```python
#Solution 3

def GetAreaOfTriangle(base,height):
    '''This funtion will take 2 integer as input(base & height) & provide result as Area of Triangle.
    '''
    try:
        area = 1/2*(base*height)
        return (area)
    except Exception as e:
        print(e)
GetAreaOfTriangle(2,5)
```

```python
#Solution 4

def Swap2Variables(a,b):
    '''This funtion will take 2 integer variable as input & swap those 2 variables values.
    '''
    try:
        temp = a
        a = b
        b = temp
        return (a,b)
    except Exception as e:
        print(e)
Swap2Variables(3,9)
```

```python
#Solution 5
import random
def GenerateRandomNum(a,b):
    '''This funtion will take 2 integer number & generate random numbers between those 2 numbers.
    '''
    try:
        return (random.randint(a,b))
    except Exception as e:
        if(b<a):
            print("Start Number is greater than End Number")
        else:
            print(e)
        
GenerateRandomNum(1,10)
```
