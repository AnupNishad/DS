```python
#Solution 1

def ConvertKMToMile(Km):
    # As 1 Kilometer = 0.62137 Miles
    '''This function will take Kilometer input & provide Miles as output.
    '''
    try:
        Miles = Km*0.62137
        return Miles
    except Exception as e:
        print(e)
        
ConvertKMToMile(2)
```

```python
#Solution 2

def ConvertCelsiusToFahrenheit(Celsius):
    # As 1 Celsius = (1 C *9/5) + 32 Fahrenheit
    '''This function will take Celsius input & provide Fahrenheit as output.
    '''
    try:
        fahrenheit = (Celsius*9/5) + 32
        return fahrenheit
    except Exception as e:
        print(e)
        
ConvertCelsiusToFahrenheit(2)
```

```python
#Solution 3
import calendar
def DisplayCalendar(yy):
    '''This function will take year input & provide calendar for that year as output.
    '''
    try:
        #print( calendar.month(yy, mm))--> for specific month calendar
        print(calendar.calendar(yy)) #for specific year calendar
    except Exception as e:
        print(e)
        
DisplayCalendar(2022)
```

```python
#Solution 5
'''This function will take 2 integer valriable input & provide swapping those 2 variables as output.
'''
def SwapNumbersNoTemp(a,b):
    try:
        a = a+b
        b = a-b
        a = a-b
        return (a,b)
    except Exception as e:
        print(e)

SwapNumbersNoTemp(4,5)
```
