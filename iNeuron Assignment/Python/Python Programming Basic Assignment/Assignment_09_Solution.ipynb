```python
#Solution 1

def CheckDisariumNumber(Number):
    '''This function will take integer input & check if its Disarium Number or not.
    '''
    try:
        Sum = 0
        OriginalNumber = Number
        NumberLength = int(len(str(Number)))
        while Number>0:
            digit = Number%10
            Sum = Sum+(digit**NumberLength)
            Number = Number//10
            NumberLength = NumberLength-1
        if Sum == OriginalNumber:
            print(f"{OriginalNumber} is a disarium number")
        else:
            print(f"{OriginalNumber} is not a disarium number")
    except Exception as e:
        print("Error : ",e)
        
CheckDisariumNumber(135)
CheckDisariumNumber(153)
```

```python
#Solution 2

def CheckDisariumNumber(Number):
    '''This function will take integer input & check if its Disarium Number or not.
    '''
    try:
        Sum = 0
        OriginalNumber = Number
        NumberLength = int(len(str(Number)))
        while Number>0:
            digit = Number%10
            Sum = Sum+(digit**NumberLength)
            Number = Number//10
            NumberLength = NumberLength-1
        if Sum == OriginalNumber:
            print(OriginalNumber,end=" ")
    except Exception as e:
        print("Error : ",e)

def DisariumNumber1To100():
    '''This function will print all Disarium Number between 1-100.
    '''
    try:
        print("Disarium Number between 1-100 are : ",end=" ")
        for i in range(1,101):
            CheckDisariumNumber(i)
    except Exception as e:
        print("Error : ",e)
        
DisariumNumber1To100()
```

```python
#Solution 3

def GetDigitSquareSum(Number):
    '''This function will take integer input & provide each digit square sum.
    '''
    try:
        Sum = 0
        while Number>0:
            digit = Number%10
            Sum = Sum+(digit**2)
            Number = Number//10
        return Sum
    except Exception as e:
        print("Error : ",e)
        
def CheckHappyNumber(Number):
    '''This function will take integer input & check if its Happy Number or not.
    '''
    try:
        isHappyNum = Number
        while(isHappyNum != 1 and isHappyNum != 4):    
            isHappyNum = GetDigitSquareSum(isHappyNum);   
        if isHappyNum == 1:
            print(f"{Number} is a Happy Number")
        else:
            print(f"{Number} is not a Happy Number")
    except Exception as e:
        print("Error : ",e)
        
CheckHappyNumber(13)
CheckHappyNumber(4)
```

```python
#Solution 4

def GetDigitSquareSum(Number):
    '''This function will take integer input & provide each digit square sum.
    '''
    try:
        Sum = 0
        while Number>0:
            digit = Number%10
            Sum = Sum+(digit**2)
            Number = Number//10
        return Sum
    except Exception as e:
        print("Error : ",e)
        
def CheckHappyNumber(Number):
    '''This function will take integer input & check if its Happy Number or not.
    '''
    try:
        isHappyNum = Number
        while(isHappyNum != 1 and isHappyNum != 4):    
            isHappyNum = GetDigitSquareSum(isHappyNum);   
        if isHappyNum == 1:
            print(Number,end=" ")
    except Exception as e:
        print("Error : ",e)

def HappyNumber1To100():
    '''This function will print all Happy Number between 1-100.
    '''
    try:
        print("Happy Number between 1-100 are : ",end=" ")
        for i in range(1,101):
            CheckHappyNumber(i)
    except Exception as e:
        print("Error : ",e)
        
HappyNumber1To100()
```

```python
#Solution 6

def CheckPronicNumber(Number):
    '''This function will take integer input & check if it's Pronic Number or not.
    '''
    try:
        for i in range(1, Number+1): 
            if((i*(i+1)) == Number): 
                print(Number,end=" ")
                break;
    except Exception as e:
        print("Error : ",e)

def PronicNumber1To100():
    '''This function will print all Pronic Number between 1-100.
    '''
    try:
        print("Pronic Number between 1-100 are : ",end=" ")
        for i in range(1,101):
            CheckPronicNumber(i)
    except Exception as e:
        print("Error : ",e)
    
PronicNumber1To100()
```
