```python
#Solution 1

def GetUniqueDictValues(Dict):
    '''This function will take a dictionary input & provide unique dictionary values.
    '''
    try:
        ListUniqueValue = []
        print("Before exctracting unique dictionary values : ",list(Dict.values()))
        for i in Dict.values():
            if i not in ListUniqueValue:
                ListUniqueValue.append(i)
        print("After exctracting unique dictionary values : ",ListUniqueValue)
    except Exception as e:
        print("Error : ",e)
        
GetUniqueDictValues({"key1":"iNeuron","key2":13,"key3":"ML","key4":51,"key5":19,"key6":13,"key7":"ML","key8":99})
```

```python
#Solution 2

def GetSumDictItems(Dict):
    '''This function will take a dictionary input & provide sum of all dictionary items.
    '''
    try:
        Sum = 0
        for i in Dict.values():
            if type(i) == int:
                Sum = Sum+i
        print("Sum of all integer dictionary items are  : ",Sum)
    except Exception as e:
        print("Error : ",e)
        
GetSumDictItems({"key1":"iNeuron","key2":13,"key3":"ML","key4":51,"key5":19,"key6":23,"key7":"ML","key8":99})
```

```python
#Solution 3

def Merge2Dictionaries(Dict1,Dict2):
    '''This function will take 2 dictionaries input & merge both into one.
    '''
    try:
        print("Dict1 : ",Dict1)
        print("Dict2 : ",Dict2)
        Dict1.update(Dict2)
        print("Merged : ",Dict1)
    except Exception as e:
        print("Error : ",e)
        
Merge2Dictionaries({"key1":"iNeuron","key2":13,"key3":"ML"},{"key4":51,"key5":19,"key6":23,"key7":"ML","key8":99})
```

```python
#Solution 4

from itertools import product
def GetFlatDictionary(Dict):
    '''This function will take a dictionary input & provide flattened dictionary.
    '''
    try:
        print("The original dictionary is : " + str(Dict))
        Flattened = dict(zip(Dict['Day'], Dict['DayName']))
        print("Flattened dictionary : " + str(Flattened))
    except Exception as e:
        print("Error : ",e)
        
GetFlatDictionary({"Day":[1,2,3,4,5,6,7],"DayName":["Monday","Tuesday","Wednessday","Thursday","Friday","Saturday","Sunday"]})
```

```python
#Solution 5

from collections import OrderedDict
def InsertItemInBeginingIntoDict(Dict):
    '''This function will take a dictionary input & insert a item in begining of the dictionary.
    '''
    try:
        NewDict = OrderedDict(Dict)
        print("Original dictionary : ",NewDict)
        NewDict.update({'key4':'4'})
        NewDict.move_to_end('key4', last = False)
        print("After Insertion : " + str(NewDict))
    except Exception as e:
        print("Error : ",e)
        
InsertItemInBeginingIntoDict([('key1','1'),('key2','2'),('key3','3')])
```

```python
#Solution 6

from collections import OrderedDict
def CheckCharOrderInDict(Dict,Char):
    '''This function will take a dictionary input & order of character string in the dictionary.
    '''
    try:
        NewDict = OrderedDict.fromkeys(Dict)
        
        CharLen = 0
        for i,j in NewDict.items(): 
            if (i == Char[CharLen]): 
                CharLen = CharLen + 1 
            if (CharLen == (len(Char))): 
                return "The order of pattern is correct"
        return "The order of pattern is not correct"
    except Exception as e:
        print("Error : ",e)
        
CheckCharOrderInDict("This is my first python program","ir")
```
