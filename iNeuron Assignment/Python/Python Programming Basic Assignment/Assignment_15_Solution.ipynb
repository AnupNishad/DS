```python
#Solution 1

def GetDivisibleNumBy5N7UsingGenerator(n):
    '''This function will take integer input & provide divisible number by 5 & 7 upto provided number using generator.
    '''
    try:
        for i in range(0,n+1):
            if (i%5==0 and i%7==0):
                yield i
    except Exception as e:
        print("Error : ",e)
        
Number = int(input("Enter Number : "))
print(f"Numbers divisible by 5 & 7 using generator function upto {Number} is/are : ",end = " ")
lis = []
for i in GetDivisibleNumBy5N7UsingGenerator(Number):
    lis.append(str(i))
print (','.join(lis))
```

```python
#Solution 2

def GetEvenNumUsingGenerator(n):
    '''This function will take integer input & provide Even number upto provided number using generator.
    '''
    try:
        for i in range(0,n+1):
            if (i%2==0):
                yield i
    except Exception as e:
        print("Error : ",e)
        
Number = int(input("Enter Number : "))
print(f"Even Numbers using generator function upto {Number} is/are : ",end = " ")
for i in GetEvenNumUsingGenerator(Number):
    if i==Number or i==Number-1:
        print(i, end = " ")
    else:
        print(i, end = ",")
```

```python
#Solution 3

def GetFibonacciSeqUsingListComprehension(n):
    '''This function will take a integer input & provide fibonacci sequence till that number using list comrehension.
    '''
    try:
        mylist = [0,1]
        [mylist.append(mylist[-2]+mylist[-1]) for i in range(n-1)]
        return (mylist)
    except Exception as e:
         print("Error : ",e)

Number = int(input("Enter Number : "))
print(f"Fibonacci sequence using list comprehension function upto {Number} is/are : ",end = " ")
GetFibonacciSeqUsingListComprehension(Number)
```

```python
#Solution 4

def GetuserNameFromEmailAddress(EmailAddress):
    '''This function will take a string input & provide username from email address provided.
    '''
    try:
        Index = EmailAddress.find("@")
        print(f"User Name from Email Address {EmailAddress} is : {EmailAddress[0:Index]}")
    except Exception as e:
         print("Error : ",e)
            
GetuserNameFromEmailAddress("john@google.com")
```
