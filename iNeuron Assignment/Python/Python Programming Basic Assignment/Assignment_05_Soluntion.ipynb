```python
#Solution 1

def GetLCM(Num1,Num2):
    '''This function will take 2 integer input & provide LCM of those numbers as output
    '''
    try: 
        if Num1>Num2:
            greater = Num1
        else:
            greater = Num2

        while True:
            if ((greater%Num1 == 0) and (greater%Num2 == 0)):
                lcm = greater
                break
            greater = greater+1
        return lcm
    except Exception as e:
        print(e)
        
GetLCM(4,3) 
```

```python
#Solution 2

def GetHCF(Num1,Num2):
    '''This function will take 2 integer input & provide HCF of those numbers as output
    '''
    try: 
        if Num1<Num2:
            smaller = Num1
        else:
            smaller = Num2

        for i in range(1,smaller+1):
            if ((Num1%i == 0) and (Num2%i == 0)):
                hcf = i
        return hcf
    except Exception as e:
        print(e)
        
GetHCF(15,20)
```

```python
#Solution 3

def ConvertDecimalToBinaryOctalHexadecimal(decimal):
    '''This function will take a integer input & provide equavalent binary,octal & hexadecimal of that numbers as output
    '''
    try:
        return bin(decimal),oct(decimal),hex(decimal)
    except Exception as e:
        print(e)

ConvertDecimalToBinaryOctalHexadecimal(20)
```

```python
#Solution 4

def GetASCIIValue(char):
    '''This function will take a character input & provide equavalent ASCII value of that char as output
    '''
    try:
        return ord(char)
    except Exception as e:
        print(e)

GetASCIIValue('A')
```
