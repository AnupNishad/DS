```python
#Solution 1

def recursive_fibonacci(Number):
    '''This function will take a integer input & provide fibonacci sequence using recursive as output
    '''
    try:
        if Number <= 1:
               return Number
        else:
            return(recursive_fibonacci(Number-1) + recursive_fibonacci(Number-2))
    except Exception as e:
        print(e)

def Call_Recursive_Fibonacci(Number):
    '''This function will take a integer input & will invoke the recursive_fibonacci() to get fibonacci sequence
    '''
    try:
        if Number < 0:
            print("Plese enter a positive integer")
        else:
            for i in range(Number+1):
                print(recursive_fibonacci(i),end=" ")
    except Exception as e:
        print(e)
            
Call_Recursive_Fibonacci(10)
```

```python
#Solution 2
def recursive_factorial(Number):
    '''This function will take a integer input & provide factorial using recursive as output
    '''
    try:
        if Number == 1:
               return Number
        else:
            return(Number*recursive_factorial(Number-1))
    except Exception as e:
        print(e)

def Call_Recursive_Factorial(Number):
    '''This function will take a integer input & will invoke the recursive_factorial() to get factorial
    '''
    try:
        if Number < 0:
            print("Plese enter a positive integer")
        else:
                print(recursive_factorial(Number))
    except Exception as e:
        print(e)
            
Call_Recursive_Factorial(10)
```

```python
#Solution 3

def GetBMI(weightinkg,heightincentimeter):
    '''This function will take 2 integer/float height(in Centimeter) & weight(in Kg) as input & provide BMI as output
    '''
    try:
        BMI = weightinkg/heightincentimeter**2
        print(BMI)
    except Exception as e:
        print(e)
            
GetBMI(60,1.67)
```

```python
#Solution 4

import math
def GetNaturalLogarithm(Number):
    '''This function will take a integer/float input & provide natural logarithm of that number as output
    '''
    try:
        return (math.log(Number))
    except Exception as e:
        print(e)
            
GetNaturalLogarithm(10)
```

```python
#Solution 5

def GetQubeSumNaturalNumber(Number):
    '''This function will take a integer input & provide cube sum of till that natural number as output
    '''
    try:
        sum = 0
        for i in range(1,Number+1):
            sum = sum+i**3
        return (sum)
    except Exception as e:
        print(e)
            
GetQubeSumNaturalNumber(10)
```
