```python
#Solution 1

def GetListElementSum(l):
    '''This function will take list input & provide summation of list element as output
    '''
    try:
        sum = 0
        for i in l:
            sum = sum+i
        return (sum)
    except Exception as e:
        print("Error : ",e)
    
GetListElementSum([1,6,9,8,11,30,7])
```

```python
#Solution 2

def GetListNumberElementMultiplication(l):
    '''This function will take list input & provide multiplication of all number element as output
    '''
    try:
        multiply = 1
        for i in l:
            if type(i) == int:
                multiply = multiply*i
        return (multiply)
    except Exception as e:
        print("Error : ",e)
    
GetListNumberElementMultiplication([1,5,6,8,"iNeuron",10,"ML"])
```

```python
#Solution 3

def GetSmallestNumberFromList(l):
    '''This function will take list input & provide smallest list number as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                l1.append(i)
        l1.sort()
        return (l1[0])
    except Exception as e:
        print("Error : ",e)
    
GetSmallestNumberFromList([17,25,6,7,8,"iNeuron",70,"ML"])
```

```python
#Solution 4

def GetLargestNumberFromList(l):
    '''This function will take list input & provide largest list number as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                l1.append(i)
        l1.sort(reverse = True)
        return (l1[0])
    except Exception as e:
        print("Error : ",e)
    
GetLargestNumberFromList([17,25,6,7,8,"iNeuron",70,"ML"])
```

```python
#Solution 5

def Get2ndLargestNumberFromList(l):
    '''This function will take list input & provide Second largest list number as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                l1.append(i)
        l1.sort(reverse = True)
        return (l1[1])
    except Exception as e:
        print("Error : ",e)
    
Get2ndLargestNumberFromList([17,25,6,7,8,"iNeuron",70,"ML"])
```

```python
#Solution 6

def GetNLargestNumberFromList(l,nLargest):
    '''This function will take list input & provide n largest list number as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                l1.append(i)
        l1.sort(reverse = True)
        return (l1[0:nLargest])
    except Exception as e:
        print("Error : ",e)
    
GetNLargestNumberFromList([17,25,6,7,8,"iNeuron",70,"ML",55,99],3)
```

```python
#Solution 7

def GetEvenNumberFromList(l):
    '''This function will take list input & provide even number from list as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                if i%2 == 0:
                    l1.append(i)
        return (l1)
    except Exception as e:
        print("Error : ",e)
    
GetEvenNumberFromList([17,25,6,7,9,1,12,8,"iNeuron",70,"ML"])
```

```python
#Solution 8

def GetOddNumberFromList(l):
    '''This function will take list input & provide odd number from list as output
    '''
    try:
        l1 = []
        for i in l:
            if type(i) == int:
                if i%2 != 0:
                    l1.append(i)
        return (l1)
    except Exception as e:
        print("Error : ",e)
    
GetOddNumberFromList([17,25,6,7,9,12,8,"iNeuron",70,"ML"])
```

```python
#Solution 10
import copy
def GetListCopy(l):
    '''This function will take list input & create a same copy of the list provided as input
    '''
    try:
        l1 = copy.copy(l)
        l2 = copy.deepcopy(l)
        print("Original List : ",l)
        print("Copy of List using copy.copy method : ",l1)
        print("Copy of List using copy.deepcopy method : ",l2)
    except Exception as e:
        print("Error : ",e)
    
GetListCopy([[],17,25,6,7,[],9,12,8,"iNeuron",70,"ML"])
```

```python
#Solution 11
import copy
def CountOccurrencesOfElementInList(l,element):
    '''This function will take list input & provide count of occurrences of element from list
    '''
    try:
        count = 0
        for i in l:
            if i == element:
                count = count+1
        return (f"Occurrence(s) of element {element} is : {count}")
    except Exception as e:
        print("Error : ",e)
    
CountOccurrencesOfElementInList([17,25,6,7,9,12,8,17,"iNeuron",70,"ML",17,8,"iNeuron"],17)
```
