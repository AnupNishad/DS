```python
#Solution 1

def GetWordsGreaterThanGivenLength(statement,length):
    '''This function will take string input & provide words greater than length provided.
    '''
    try:
        for i in statement.split(" "):
            if len(i)>length:
                print(i,end=" ")
    except Exception as e:
        print("Error : ",e)

GetWordsGreaterThanGivenLength("This is my first python program",4)  
```

```python
#Solution 2

def RemoveithCharFromString(String,CharPosition):
    '''This function will take string input & remove character position provided.
    '''
    try:
        StringWithoutithChar = ""
        for i in range(len(String)):
            if i != CharPosition:
                StringWithoutithChar = StringWithoutithChar+String[i]
        print(StringWithoutithChar)
    except Exception as e:
        print("Error : ",e)

RemoveithCharFromString("This is my first python program",17)
```

```python
#Solution 3


def SplitAndJoinString(String):
    '''This function will take string input & will split & join string based on delimiter provided.
    '''
    try:
        SplitString = String.split(" ")
        JoinString = "-".join(SplitString)
        print("String splitted based on space(" ") delimiter : ",SplitString)
        print( "String joined based on Hyphen(-) delimiter : ",JoinString)
    except Exception as e:
        print("Error : ",e)

SplitAndJoinString("This is my first python program")
```

```python
#Solution 4

def CheckBinaryString(String):
    '''This function will take string input & check if its binary or not.
    '''
    try:
        BinarySet = {'0','1'}
        ConvertedString = set(String)
        
        if BinarySet == ConvertedString or ConvertedString == {'0'} or ConvertedString == {'1'}:
            print(String,"is a binary string.")
        else:
            print(String,"is not a binary string.")
    except Exception as e:
        print("Error : ",e)

CheckBinaryString("Hello World")
CheckBinaryString("111")
```

```python
#Solution 5

def FindUncommonWordsFrom2String(String1,String2):
    '''This function will take 2 string input & provide uncommon string from both string.
    '''
    try:
        String1 = String1.split()
        String2 = String2.split()
        print("First String : ",String1)
        print("Second String : ",String2)
        UncommonWord = list(set(String1).symmetric_difference(set(String2)))
        print("Uncommon Word : ",UncommonWord)
    except Exception as e:
        print("Error : ",e)

FindUncommonWordsFrom2String("Hello World","Hello Python World")
```

```python
#Solution 6

def DuplicateCharactersFromString(String):
    '''This function will take string input & provide duplicate characters from string.
    '''
    try:
        DuplicatesList = []
        for i in String:
            if String.count(i) > 1:
                    if i not in DuplicatesList:
                           DuplicatesList.append(i)
        print(f"Duplicate Character in {String} is/are  {DuplicatesList}")
    except Exception as e:
        print("Error : ",e)

DuplicateCharactersFromString("Hello World")
```

```python
#Solution 7

import re
def CheckStringContainSpecialChar(String):
    '''This function will take string input & check if string contains special character or not.
    '''
    try:
        SpecialChar = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
        if(SpecialChar.search(String) == None):
            print(f"{String} not contains special character")
          
        else:
            print(f"{String} contains special character")
    except Exception as e:
        print("Error : ",e)

CheckStringContainSpecialChar("Hello World")
CheckStringContainSpecialChar("H@llo World")
```
