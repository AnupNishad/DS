```python
#Solution 1

def CheckNumber(Number):
    '''This funtion will take a number input & provide output as whether Number is Positive, Negative or Zero
    '''
    try:
        if Number == 0:
            return "Provided Number is Zero(0)"
        elif Number < 0:
            return "Provided Number is a Negative Number"
        else:
            return "Provided Number is a Positive Number"
    except Exception as e:
        print(e)

CheckNumber(0)
CheckNumber(1)
CheckNumber(-1)
```

```python
#Solution 2

def CheckOddEven(Number):
    '''This funtion will take a number input & provide Number is Odd or Even as output
    '''
    try:
        if Number%2 == 0:
            return "Provided Number is a Even Number"
        else:
            return "Provided Number is a Odd Number"
    except Exception as e:
        print(e)

CheckOddEven(5)
CheckOddEven(6)
```

```python
#Solution 3

def CheckLeapYear(Year):
    '''This funtion will take a year input & provide year is leap or not leap as output
    '''
    try:
        if (Year%4 == 0 and Year%100 != 0 or Year%400 == 0):
            return "Provided Year is a Leap Year"
        else:
            return "Provided Year is not a Leap Year"
    except Exception as e:
        print(e)

CheckLeapYear(2022)
CheckLeapYear(2024)
```

```python
#Solution 4

def CheckPrimeNumber(Number):
    '''This funtion will take a number input & provide number is prime or not prime as output
    '''
    try:
        if Number > 1:
                for i in range(2, int(Number/2)+1):
                    if (Number % i) == 0:
                        print("Number is not a prime number")
                        break
                else:
                    print("Number is a prime number")
        else:
            print("Number is not a prime number")
    except Exception as e:
        print(e)
        
CheckPrimeNumber(29)
```
