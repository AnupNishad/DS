```python
#Solution 3

'''
     __init__ method of the child class overrides the __init__ method of the parent class.
     Therefore we have to manually call a parent superclass's __init__ method using super() method.
    
    E.g : 
        class Student:
            def __init__(self,name,age,gender):
                self.name = name
                self.age = age 
                self.gender = gender 
        class Girl(Student):
            def __init__(self,name,age,gender,subject,marks):
                super().__init__(name,age,gender)
                self.subject = subject
                self.marks = marks
                
        girl1 = Girl('Gayathri',28,"F","English",70)
        print(girl1.__dict__)
        
        Output : 
        --------------------------------------------------------------------------------
        {'name': 'Gayathri', 'age': 28, 'gender': 'F', 'subject': 'English', 'marks': 70}
         
'''
```

```python
#Solution 4

'''
    super() method can be used to augment, instead of completely replacing, an inherited method.
    
    E.g : 
        class Student:
            def __init__(self,name,age,gender):
                self.name = name
                self.age = age 
                self.gender = gender 
        class Girl(Student):
            def __init__(self,name,age,gender,subject,marks):
                super().__init__(name,age,gender)
                self.subject = subject
                self.marks = marks
                
        girl1 = Girl('Gayathri',28,"F","English",70)
        print(girl1.__dict__)
        
        Output : 
        --------------------------------------------------------------------------------
        {'name': 'Gayathri', 'age': 28, 'gender': 'F', 'subject': 'English', 'marks': 70}
    
'''
```
