```python
#Solution 2

'''
    We can create class instance by calling a class by its name and pass the arguments which its __init__ method accepts.
    
    We can create a class by using class keyword followed by class name and semicolon (:).
    
    
    e.g : 
        
        class student : 
            def __init__(self,rollno,subject,marks):
                    self.rollno = rollno
                    self.subject = subject
                    self.marks = marks
        
        Here student is the name of class created with class keyword with __init__() method.
    
    e.g : 
        
        akshay = student(101,"Maths",75)
        Here akshay is an instance of class student with attributes 101 (Roll No.),"Maths" (Subject) and 75 (Marks).    
    
'''
```

```python
#Solution 3

'''
    class attributes created at top of class definiation outside of all methods.
    
    e.g : 
         class student : 
             class_division = "A" 
            def __init__(self,rollno,subject,marks):
                self.rollno = rollno
                self.subject = subject
                self.marks = marks  
                
        Here class_division = "A" is a class attributes
    
'''
```

```python
#Solution 4

'''
    instance attributes created inside __init__() method.
    
    e.g : 
         class student : 
             class_division = "A" 
            def __init__(self,rollno,subject,marks):
                self.rollno = rollno   #|
                self.subject = subject #| These are instance attributes
                self.marks = marks     #|
                
         student1 = student(101,"Maths",75)
         student2 = student(102,"Maths",60)
         
        Here student1 & student2 are both the instances of class student with different instance variables.
    
'''
```

```python
#Solution 5

'''
    The term "self" in a Python class mean pointer inside __init__() method.
    
    e.g : 
         class student : 
             class_division = "A" 
            def __init__(self,rollno,subject,marks):
                self.rollno = rollno   #|
                self.subject = subject #| These are instance attributes
                self.marks = marks     #|
                
         student1 = student(101,"Maths",75)
         student2 = student(102,"Maths",60)
         
         print(student1.__dict__)
         print(student2.__dict__)
         
         Output :
         -----------------------------
         {'rollno': 101, 'subject': 'Maths', 'marks': 75}
         {'rollno': 102, 'subject': 'Maths', 'marks': 60}
         
        Here when we are creating student1 & student2 instances of class student with different instance variables those values are passing to the __init__() method for processing.
    
'''
```

```python
#Solution 8

'''
    The most popular form of operator overloading in python is by special methods.
    
    e.g : 
        class AddOperatoroverloading:
            def __init__(self,a):
                self.a = a
            def __add__(self,b):
                return self.a+b.a
    
        obj1 = AddOperatoroverloading(1)
        obj2 = AddOperatoroverloading(2)
        obj3 = AddOperatoroverloading('Python')
        obj4 = AddOperatoroverloading(' Program')
        print(obj1+obj2)
        print(obj3+obj4)
        
        Output : 
        -----------------------------------
        3
        Python Program
    
'''
```
