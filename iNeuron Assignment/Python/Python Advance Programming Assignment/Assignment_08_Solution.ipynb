```python
#Solution 1

def vowel_links(String):
    '''
        This function will take a string input & provides True if any two adjacent words have this
        property: One word ends with a vowel, while the word immediately after begins with a vowel (a e i o u).
    '''
    try:
        StringSplitBySpace = String.split(" ")
        Vowel = ["a","e","i","o","u"]
        Start = False
        End = False
        StringStartEndsWithVowel = False
        for i in range(len(StringSplitBySpace)):
            if (StringSplitBySpace[i][-1] in Vowel):
                EndCheck = True
            else:
                EndCheck = False
            
            if (StringSplitBySpace[i][0] in Vowel):
                Start = True
            else:
                Start = False
            if (Start == True and End == True):
                StringStartEndsWithVowel = True
                break
            End = EndCheck
            
        print(f'vowel_links({String}) ➞ {StringStartEndsWithVowel}')
        
    except Exception as e:
        print("Error : ",e)
        
vowel_links("a very large appliance")
vowel_links("go to edabit")
vowel_links("an open fire")
vowel_links("a sudden applause")
```

```python
#Solution 2

def first_before_second(String,Letter1,Letter2):
    '''
        This function will take 3 input(a string,one letter,second letter) & provides True if every 
        instance of the first letter occurs before every instance of the second letter.
    '''
    try:
        first_before_second = True
        HighestIndexOfLetter1 = String.rindex(Letter1)
        LowestIndexOfLetter2 = String.index(Letter2)
        
        if HighestIndexOfLetter1 > LowestIndexOfLetter2 :
            first_before_second = False
        
        print(f'first_before_second({String}, {Letter1}, {Letter2}) ➞ {first_before_second}')
        
    except Exception as e:
        print("Error : ",e)
        
first_before_second("a rabbit jumps joyfully", "a", "j")
first_before_second("knaves knew about waterfalls", "k", "w")
first_before_second("happy birthday", "a", "y")
first_before_second("precarious kangaroos", "k", "a")
```

```python
#Solution 3

def char_at_pos(OddEvenList,OddEevenSpecifier):
    '''
        This function will take 2 input(a List,string) & provides the characters from a list or string r on odd or
        even positions, depending on the specifier s. The specifier will be "odd" for
        items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2,4, 6, ...).
    '''
    try:
        OddOREvenList = []
        for i in range(0,len(OddEvenList)):
            #print(i,OddEvenList[i],i+1)
            if OddEevenSpecifier == "even":
                if (i+1)%2 == 0:
                    OddOREvenList.append(OddEvenList[i])
            else:
                if (i+1)%2 != 0:
                    OddOREvenList.append(OddEvenList[i])
        if type(OddEvenList) == str:
            OddOREvenList = "".join(OddOREvenList)
        print(f'char_at_pos({OddEvenList}, {OddEevenSpecifier}) ➞ {OddOREvenList}')
        
    except Exception as e:
        print("Error : ",e)
        
char_at_pos([2, 4, 6, 8, 10], "even")
char_at_pos("EDABIT", "odd")
char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd")
```

```python
#Solution 4

def GCD(GCDList):
    '''
        This function will take a input(a List) & provides the greatest common divisor of all list
        elements. If the greatest common divisor is 1, return 1.
    '''
    try:
        Min = min(GCDList)
        GCD = 0
        for i in range(1, Min+1):
            GCDElement = []
            for j in GCDList:
                GCDElement.append(j%i)
            if len(set(GCDElement)) == 1 and list(set(GCDElement))[0] == 0:
                GCD = i
        print(f'GCD({GCDList}) ➞ {GCD}')
        
    except Exception as e:
        print("Error : ",e)
        
GCD([10, 20, 40])
GCD([1, 2, 3, 100])
GCD([1024, 192, 2048, 512])
```

```python
#Solution 5

def palindrome_type(Number):
    '''
        This function will take a input(a List) & provides if n or the binary representation of n is palindromic. 
        Return the following:
        "Decimal only." if only n is a palindrome.
        "Binary only." if only the binary representation of n is a palindrome.
        "Decimal and binary." if both are palindromes.
        "Neither!" if neither are palindromes.
    '''
    try:
        BinaryNumber  = bin(Number).replace("0b","")
        ReverseBinaryNumber = str(BinaryNumber)[::-1]
        
        StringNumber = str(Number)
        ReverseStringNumber  = StringNumber[::-1]
        
        Msg = ""
        
        if (StringNumber == ReverseStringNumber) and (BinaryNumber == ReverseBinaryNumber):
            Msg = "Decimal and binary."
        
        elif (BinaryNumber == ReverseBinaryNumber):
            Msg = "Binary only."
            
        elif (StringNumber == ReverseStringNumber):
            Msg = "Decimal only."
            
        else:
            Msg = "Neither!"
        
        print(f'palindrome_type({Number}) ➞ {Msg}')
        
    except Exception as e:
        print("Error : ",e)
        
palindrome_type(1306031)
palindrome_type(427787)
palindrome_type(313)
palindrome_type(934)
```
