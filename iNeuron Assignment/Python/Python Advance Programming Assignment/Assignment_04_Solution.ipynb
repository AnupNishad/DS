```python
#Solution 1

def fib_fast(Number):
    '''
        This function will take integer input & provides Fibonacci Number for given number.
    '''
    try:
        f1=0
        f2=1
        for x in range(2,Number+1):
            nextsequence=f1+f2
            f1=f2
            f2=nextsequence
        print(f" fib_fast({Number}) ➞ {nextsequence}")
    except Exception as e:
        print("Error : ",e)
        
fib_fast(5)
fib_fast(10)
fib_fast(20)
fib_fast(50)
```

```python
#Solution 2

def convert_to_hex(String):
    '''
        This function will take string input & provides equavalent hexadecimal number.
    '''
    try:
        HexString = ""
        for i in String:
            HexString = HexString+hex(ord(i))
        NewHexString = HexString.replace("0x"," ")
        print(f"convert_to_hex({String}) ➞ {NewHexString}")
    except Exception as e:
        print("Error : ",e)
        
convert_to_hex("hello world")
convert_to_hex("Big Boi")
convert_to_hex("Marty Poppinson")
```

```python
#Solution 3

def uncensor(CensoredString,CensoredVowels):
    '''
        This function will take 2 string input & provides output after replacing * with appropriate vowels.
    '''
    try:
        VowelIndex = 0
        UncensoredString = ""
        for i in CensoredString:
            if i == "*":
                UncensoredString = UncensoredString+CensoredVowels[VowelIndex]
                VowelIndex = VowelIndex+1
            else:
                UncensoredString = UncensoredString+i
        print(f"uncensor({CensoredString},{CensoredVowels}) ➞ {UncensoredString}")
    except Exception as e:
        print("Error : ",e)
        
uncensor("Wh*r* d*d my v*w*ls g*?", "eeioeo")
uncensor("abcd", "")
uncensor("*PP*RC*S*", "UEAE") 
```

```python
#Solution 4

import socket

def get_domain(IPAddress):
    '''
        This function will take a string input & provides the domain name using PTR DNS records.
    '''
    try:
        print(f"get_domain({IPAddress}) ➞ {list(socket.gethostbyaddr(IPAddress))[0]}")
    except Exception as e:
        print("Error : ",e)
        
get_domain("8.8.8.8")
get_domain("8.8.4.4")
```

```python
#Solution 5

def fact_of_fact(Number):
    '''
        This function will take a integer input & provides the factorial of factorial of given number.
    '''
    def factorial(Num):
            '''
                This function will take a integer input & provides the factorial of given number.
            '''
            try:
                j = 1
                for i in range(Num,0,-1):
                    j = j*i
                return j
            except Exception as e:
                print("Error : ",e)
    fact_of_fact = 1
    for k in range(1,Number+1):
        fact_of_fact = fact_of_fact*factorial(k)
    print(f"fact_of_fact({Number}) ➞ {fact_of_fact}")
    
fact_of_fact(4)
fact_of_fact(5)
fact_of_fact(6)
```
