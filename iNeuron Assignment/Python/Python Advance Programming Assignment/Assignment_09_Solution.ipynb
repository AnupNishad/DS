```python
#Solution 1

def playback_duration(VideoDuration,PlaybackSpeed):
    '''
        This function take 2 input(video duration,playback speed) & provide the 
        actual duration and playback speed of the video, calculate the playback duration of the video.
    '''
    try:
        Time = VideoDuration.split(':')
        
        TimeInSeconds = (3600*int(Time[0])+60*int(Time[1])+int(Time[2]))/PlaybackSpeed
        
        TimeInHrs = str(int(TimeInSeconds/3600)) if TimeInSeconds > 3600 else '00'
        TimeInMins = str(int((TimeInSeconds%3600)/60)) if (TimeInSeconds)%3600 > 60 else '00'
        TimeInSecs = str(int((TimeInSeconds%3600)%60)) if ((TimeInSeconds)%3600)%60 > 0 else '00'
        
        print(f'playback_duration{VideoDuration,PlaybackSpeed}  ➞ {TimeInHrs}:{TimeInMins}:{TimeInSecs}')
    except Exception as e:
        print('Error Occurred : ',e)
        
playback_duration("00:30:00", 2)
playback_duration("01:20:00", 1.5)
playback_duration("51:20:09", 0.5)

```

```python
#Solution 2

def pile_of_cubes(TotalVolume):
    '''
        This function take a input(Total volume of a building) & provide the 
        find the number of cubes n required for the building & None if there is no such number.
    '''
    try:
        NewVolume = 0
        n = 0
        for i in range(1,TotalVolume):
            NewVolume = NewVolume+pow(i,3)
            if TotalVolume <= NewVolume: 
                if TotalVolume == NewVolume:
                    n = i
                else:
                    n = None
                break
        print(f'pile_of_cubes({TotalVolume}) ➞ {n}')
    except Exception as e:
        print('Error Occurred : ',e)
        
pile_of_cubes(1071225)
pile_of_cubes(4183059834009)
pile_of_cubes(16)

```

```python
#Solution 3

def find_fulcrum(FulcrumList):
    '''
        This function take a input(Fulcrum List) & provide the fulcrum of a list.
    '''
    try:
        LeftSum = 0
        RightSum = 0
        FulcrumNum = -1
        for i in range(len(FulcrumList)):
            if i==0:
                pass
            else:
                LeftSum = sum(FulcrumList[i-1::-1])
                RightSum = sum(FulcrumList[i+1:])
                if LeftSum == RightSum:
                    FulcrumNum = FulcrumList[i]
                    break
        
        print(f'find_fulcrum({FulcrumList}) ➞ {FulcrumNum}')
    except Exception as e:
        print('Error Occurred : ',e)
        
find_fulcrum([3, 1, 5, 2, 4, 6, -1])
find_fulcrum([1, 2, 4, 9, 10, -10, -9, 3])
find_fulcrum([9, 1, 9])
find_fulcrum([7, -1, 0, -1, 1, 1, 2, 3])
find_fulcrum([8, 8, 8, 8])

```

```python
#Solution 4

def sock_merchant(SockColor):
    '''
        This function take a input(Sock Color List) & returns an integer representing the number of matching
        pairs of socks that are available.
    '''
    try:
        MatchSock = 0
        SockList = []
        MatchSockList = []
        for i in SockColor:
                SockList.append([i,SockColor.count(i)//2])
                for j in SockList:
                     if j not in MatchSockList:
                        MatchSockList.append(j)

        for k in MatchSockList:
            MatchSock = MatchSock+k[1]
        
        print(f'sock_merchant({SockColor}) ➞ {MatchSock}')
    except Exception as e:
        print('Error Occurred : ',e)
        
sock_merchant([10, 20, 20, 10, 10, 30, 50, 10, 20])
sock_merchant([50, 20, 30, 90, 30, 20, 50, 20, 90])
sock_merchant([])
```

```python
#Solution 5

import re
def negative_sum(StringWithNum):
    '''
        This function take a input(string with integer) & return the sum of the negative integers only.
    '''
    try:
        result = [int(j) for j in re.findall(r'-?\d+', StringWithNum)]
        Sum = 0
        for i in result:
            if i<0:
                Sum = Sum+i
        print(f'negative_sum({StringWithNum}) ➞ {Sum}')
    except Exception as e:
        print('Error Occurred : ',e)
        
negative_sum("-12 13%14&-11")
negative_sum("22 13%14&-11-22 13 12")
```
