```python
#Solution 1

def filter_primes(List):
    '''This function takes a list input & returns a new list containing only prime numbers.
    '''
    try:
        PrimeList = []
        for i in List:
            if i > 1:
                for j in range(2, i):
                    if (i % j) == 0:
                           break
                else:
                    PrimeList.append(i)
        print(f'filter_primes({List}) ➞ {PrimeList}')
    except Exception as e:
        print('Error Occured : ',e)
        
filter_primes([7, 9, 3, 9, 10, 11, 27])
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70])
filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093,
1097])
```

```python
#Solution 2

def pop(List):
    '''This function takes a list input & returns the state after the balloon is popped.
    '''
    try:
        Balloon = []
        for i in range(len(List)):
            if List[i] != 0:
                for j in range(1,List[i]):
                    Balloon.append(j)
                for k in range(List[i],-1,-1):
                    Balloon.append(k)
            elif (i == 0):
                Balloon.append(i)
        print(f'pop({List}) ➞ {Balloon}')
    except Exception as e:
        print('Error Occured : ',e)
        
pop([0, 0, 0, 0, 4, 0, 0, 0, 0])
pop([0, 0, 0, 3, 0, 0, 0])
pop([0, 0, 2, 0, 0])
pop([0])
```

```python
#Solution 3

def loves_me(Num):
    '''This function takes a integer input(a number of petals) & return a string which repeats the phrases "Loves me" and "Loves me not" 
    for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.
    '''
    try:
        Phrase = "Loves me, loves me not"
        Phrase0 = Phrase.split(", ")[0]
        Phrase1 = Phrase.split(", ")[1]
        NewString = ""
        for i in range(1,Num+1):
            if i == Num:
                Phrase0 = Phrase0.upper()
                Phrase1 = Phrase1.upper()
            if i == 1:
                NewString = Phrase0 if NewString == "" else NewString+', '+Phrase0
            else:
                if i%2 == 0:
                    NewString = Phrase1 if NewString == "" else NewString+', '+Phrase1
                else:
                    NewString = Phrase0 if NewString == "" else NewString+', '+Phrase0
        print(f'loves_me({Num}) ➞ {NewString}')
    except Exception as e:
        print('Error Occured : ',e)
        
loves_me(3)
loves_me(6)
loves_me(1)
```

```python
#Solution 4

def sort_by_letter(List):
    '''This function takes a List input & sorts each string in a list by the letter in alphabetic ascending order (a-z).
    '''
    try:
        SortedList = []
        AlphaList = []
        for i in List:
            for j in i:
                if j.isalpha():
                    AlphaList.append(j)
                #if (ord(j)>=65 and ord(j)<=90) or (ord(j)>=97 and ord(j)<=122):
                    #SortedList.append(j)
        AlphaList.sort()
        for i in AlphaList:
            for j in List:
                if i in j:
                    SortedList.append(j)
        print(f'sort_by_letter({List}) ➞ {SortedList}')
    except Exception as e:
        print('Error Occured : ',e)
        
sort_by_letter(["932c", "832u32", "2344b"])
sort_by_letter(["99a", "78b", "c2345", "11d"])
sort_by_letter(["572z", "5y5", "304q2"])
sort_by_letter([])
```

```python
#Solution 5

def cup_swapping(List):
    '''This function takes a List input & returns the letter position that the ball is at, once I finish swapping the cups.
    '''
    try:
        BallStartPosition = "B"
        for i in List:
            if BallStartPosition in i:
                BallStartPosition = i.replace(BallStartPosition,"")
                
        print(f'sort_by_letter({List}) ➞ {BallStartPosition}')
    except Exception as e:
        print('Error Occured : ',e)
        
cup_swapping(["AB", "CA", "AB"])
```
