```python
#Solution 1


def show_the_love(List):
    
    '''This function will take a List input(List of Numbers) & it that removes 25% from every
       number in the list except the smallest number, and adds the total amount removed to the smallest number.
    '''
    try:
        MinIndex = 0
        Min = List[0]
        Element25Percent = 0
        
        for i in range(len(List)):
            if List[i] < Min:
                MinIndex = i
        
        for i in range(len(List)):
            if i != MinIndex:
                Element25Percent = Element25Percent+(List[i]*25)/100
                List[i] = List[i] - (List[i]*25)/100
        List[MinIndex] = List[MinIndex]+Element25Percent
        
        print(f'show_the_love({List}) ➞ {List}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
show_the_love([4, 1, 4])
show_the_love([16, 10, 8])
show_the_love([2, 100])
```

```python
#Solution 2


def letters(Word1,Word2):
    
    '''This function will take two words as input & returns a list of three elements, in the following order:
        1.Shared letters between two words.
        2.Letters unique to word 1.
        3.Letters unique to word 2.
    '''
    try:
        Result = []
        
        Word1Set = set(Word1)
        Word2Set = set(Word2)
        
        Result.append(''.join(sorted(Word1Set.intersection(Word2Set))))
        Result.append(''.join(sorted(Word1Set.difference(Word2Set))))
        Result.append(''.join(sorted(Word2Set.difference(Word1Set))))
        
        print(f'letters{Word1,Word2} ➞ {Result}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
letters("sharp", "soap")
letters("board", "bored")
letters("happiness", "envelope")
letters("kerfuffle", "fluffy")
letters("match", "ham")
```

```python
#Solution 3


def pairs(List):
    
    '''This function will take a list input & pairs the first number in an array with the last, the
       second number with the second to last, etc.
    '''
    try:
        Result = []
        Count = 1
        for i in range(len(List)):
            if [List[len(List)-Count],List[i]] not in Result:
                Result.append([List[i],List[len(List)-Count]])
            if Count != len(List):
                Count = Count+1
        
        
        print(f'pairs({List}) ➞ {Result}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
pairs([1, 2, 3, 4, 5, 6, 7])
pairs([1, 2, 3, 4, 5, 6])
pairs([5, 9, 8, 1, 2])
pairs([])
```

```python
#Solution 4


def add_str_nums(Str1,Str2):
    
    '''This function will take 2 string input & return addition of those 2 numbers however those numbers can be strings.
    '''
    try:
        
        DupStr1 = "0" if Str1 == "" else Str1 
        DupStr2 = "0" if Str2 == "" else Str2
        
        if DupStr1.isdigit() == False or DupStr2.isdigit() == False:
            Result = str(-1)
        else:
            Result = str(int(DupStr1)+int(DupStr2))
        
        print(f'add_str_nums{Str1,Str2} ➞ {Result}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
add_str_nums("4", "5")
add_str_nums("abcdefg", "3")
add_str_nums("1", "")
add_str_nums("1874682736267235927359283579235789257", "32652983572985729")
```

```python
#Solution 5


def unmix(Str):
    
    '''This function will take a string input & return string back to normal after swapping every pair of characters.
    '''
    try:
        NewStr = ""
        Counter = 1
        for i in range(len(Str)):
            if len(Str)%2 == 0:
                if Counter%2 == 0:
                    NewStr = NewStr+Str[i]+Str[i-1]
            else:
                if i == len(Str)-1:
                    NewStr = NewStr+Str[i]
                else:
                    if Counter%2 == 0:
                        NewStr = NewStr+Str[i]+Str[i-1]
            Counter = Counter+1
        
        print(f'unmix({Str}) ➞ {NewStr}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
unmix("123456")
unmix("hTsii  s aimex dpus rtni.g ")
unmix("badce")
```
