```python
#Solution 1


def anagram(Str,List):
    
    '''This function will take 2 input(string,list) & returns True if a given name can generate an array of words.
    '''
    try:
        StrLen = len(Str)-1
        for i in Str.split():
            for j in i:
                LengthListElement = 0
                for k in List:
                    LengthListElement = LengthListElement+len(k)
        print(f'anagram({Str}) ➞ {StrLen == LengthListElement}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
anagram("Justin Bieber", ["injures", "ebb", "it"])
anagram("Natalie Portman", ["ornamental", "pita"])
anagram("Chris Pratt", ["chirps", "rat"])
anagram("Jeff Goldblum", ["jog", "meld", "bluffs"])
```

```python
#Solution 2


def leaderboards(List):
    
    '''This function will take a list input (list) & it sorts the array to form the correct leaderboard.
    '''
    try:
        Dict1 = {}
        Result = []
        for i in range(len(List)):
            Dict1[List[i]["score"]+List[i]["reputation"]*2] = List[i]
            
        for j in sorted (Dict1.keys(),reverse=True) :
            Result.append(Dict1[j])
                    
        print(f'leaderboards({List}) ➞ {Result}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
leaderboards([
  { "name": "a", "score": 100, "reputation": 20 },
  { "name": "b", "score": 90, "reputation": 40 },
  { "name": "c", "score": 115, "reputation": 30 },
])
```

```python
#Solution 3


def hangman(String,ListOfLetters):
    
    '''This function will take 2 inputs (String & list of letters) & returns a string with hyphens - for every letter 
    of the phrase not guessed, and each letter guessed in place.
    '''
    try:
        Result = "" 
        for i in String:
            if (i.lower() in ListOfLetters or i in '"! '):
                Result = Result+i
            else:
                Result = Result+"-"
        print(f'hangman{String,ListOfLetters} ➞ {Result}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
hangman("helicopter", ["o", "e", "s"])
hangman("tree", ["r", "t", "e"])
hangman("Python rules", ["a", "n", "p", "r", "z"])
hangman("He\"s a very naughty boy!", ["e", "a", "y"])
```

```python
#Solution 4


def max_collatz(Num):
    
    '''This function will take a integer inputs & returns the highest integer in the corresponding Collatz sequence.
    '''
    try:
        OrgNum = Num
        MaxCollatz = []
        MaxCollatz.append(Num)
        while (Num !=1):
            if Num%2 == 0:
                Num = Num/2
            else:
                Num  = (Num*3)+1
                
            MaxCollatz.append(int(Num))
            if Num ==1:
                break
                
        print(f'max_collatz({OrgNum}) ➞ {max(MaxCollatz)}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
max_collatz(10)
max_collatz(32)
max_collatz(85)
```

```python
#Solution 5


def digit_sort(List):
    
    '''This function will take a List inputs & it sorts a list of integers by their digit length in
       descending order, then settles ties by sorting numbers with the same digit length in ascending order.
    '''
    try:
        MaxLen = len(str(max(List)))
        Result = []
        for i in range(MaxLen,0,-1):
            TempSort = []
            for j in List:
                if len(str(j)) == i:
                    TempSort.append(j)
            Result.extend(sorted(TempSort))
               
        print(f'digit_sort({List}) ➞ {Result}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
digit_sort([77, 23, 5, 7, 101])
digit_sort([1, 5, 9, 2, 789, 563, 444])
digit_sort([53219, 3772, 564, 32, 1])
```
