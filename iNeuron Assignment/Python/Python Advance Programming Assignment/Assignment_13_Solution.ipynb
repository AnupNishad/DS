```python
#Solution 1

def remove_letters(List,String):
    
    '''This function will take 2 input(List,String) & will remove the letters in the string from the list, and return the list.
    '''
    try:
        OriginalList = List.copy()
        for i in String:
            if i in List:
                List.remove(i)  
        print(f'remove_letters({OriginalList},{String}) ➞ {List}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
remove_letters(["s", "t", "r", "i", "n", "g", "w"], "string")
remove_letters(["b", "b", "l", "l", "g", "n", "o", "a", "w"], "balloon")
remove_letters(["d", "b", "t", "e", "a", "i"], "edabit")
```

```python
#Solution 2

def blocks(Num):
    
    '''This function will take a integer input(step) & returns the amount of blocks in that step.
    '''
    try:
        Depth = Num*3+(Num-1)*1
        for i in range(2,Num+2):
            Depth = Depth+i
        
        print(f'blocks({Num}) ➞ {Depth}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
blocks(1)
blocks(5)
blocks(2)
```

```python
#Solution 3

from operator import sub

def my_sub(Num1,Num2):
    
    '''This function will take 2 integer input(Num1,Num2) & subtracts one positive integer from another, without
       using any arithmetic operators such as -, %, /, +, etc.
    '''
    try:
        if Num1 >= Num2:
            Result = sub(Num1,Num2)
        else:
            Result = sub(Num2,Num1)
            
        
        print(f'my_sub({Num1},{Num2}) ➞ {Result}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
my_sub(5, 9)
my_sub(10, 30)
my_sub(0, 0)
```

```python
#Solution 4


def add_bill(String):
    
    '''This function will take a string input(string containing money) & returns the sum of dollar bills only, as an integer.
    '''
    try:
        DollarList = []
        for i in String.split(","):
            if ("d" and "k") in i:
                DollarList.append((int(i.replace("d","").replace("k","000"))))
            elif "d" in i:
                DollarList.append((int(i.replace("d",""))))
        
        print(f'add_bill({String}) ➞ {sum(DollarList)}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
add_bill("d20,p40,p60,d50")
add_bill("p30,d20,p60,d150,p360")
add_bill("p30,d2k,p60,d200,p360")
```

```python
#Solution 5


def flip_list(List):
    
    '''This function will take a List input(Horizontal/Vertical) & it flips a horizontal list into a vertical list, 
       and a vertical list into a horizontal list.
    '''
    try:
        Horizontal_VerticalList = []
        for i in List:
            if type(i) != list:
                Horizontal_VerticalList.append([i])
            else:
                for j in i:
                    Horizontal_VerticalList.append(j)
        
        print(f'flip_list({List}) ➞ {Horizontal_VerticalList}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
flip_list([1, 2, 3, 4])
flip_list([[5], [6], [9]])
flip_list([])
```
