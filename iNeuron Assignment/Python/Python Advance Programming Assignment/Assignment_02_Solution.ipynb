```python
#Solution 1

def GetDotsPentagon(NumIteration):
    '''
        This Function will take integer input & provides the dots exist in a pentagonal shape around the center dot on the Nth iteration.
    '''
    try:
        Pentagon = 1
        if NumIteration>0:
            for i in range(NumIteration):
                Pentagon = Pentagon+(i*5)
            print(Pentagon)
        else:
            print("Please enter a positive number")
    except Exception as e:
        print("Error : ",e)
        
GetDotsPentagon(1)
GetDotsPentagon(2)
GetDotsPentagon(3)
GetDotsPentagon(8)
```

```python
#Solution 2

def Encrypt(StringInput):
    '''
        This Function will take string input & provides the encrypted word with steps given in question.
    '''
    try:
        Vowel = {"a":0,"e":1,"i":2,"o":2,"u":3}
        ReverseString = StringInput[::-1]
        StringReplaceWithVowel = ""
        for i in ReverseString:
            if i in Vowel.keys():
                StringReplaceWithVowel = StringReplaceWithVowel+str(Vowel[i])
            else:
                StringReplaceWithVowel = StringReplaceWithVowel+str(i)
        print(StringReplaceWithVowel+"aca")
    except Exception as e:
        print("Error : ",e)
        
Encrypt("banana")
Encrypt("karaca")
Encrypt("burak")
Encrypt("alpaca")
```

```python
#Solution 3
import datetime

def has_friday_13(month,year):
    '''
        This Function will take 2 integer input(month & year) & provides whether that month contains a Friday 13th.
    '''
    try:
        if datetime.datetime(year,month,13).strftime('%A') == 'Friday':
            print("True")
        else:
            print("False")
    except Exception as e:
        print("Error : ",e)
        
has_friday_13(3,2020)
has_friday_13(10,2017)
has_friday_13(1,1985)
```

```python
#Solution 5

def pluralize(ListValue):
    '''
        This Function will take a list input & provides a set of those words in the
        plural form if they appear more than once in the list.
    '''
    try:
        FinalSet = set()
        for i in set(ListValue):
            if (ListValue.count(i) > 1 and i.endswith("y")):
                FinalSet.add(i[0:len(i)-1]+'ies')
            elif ListValue.count(i) > 1:
                FinalSet.add(i+'s')
            else:
                FinalSet.add(i)

        print(FinalSet)
    except Exception as e:
        print("Error : ",e)
        
pluralize(["cow", "pig", "cow", "cow"])
pluralize(["table", "table", "table"])
pluralize(["A", "B", "C"])
pluralize(["fly", "try", "Can","fly","try"])
```
