```python
#Solution 1

def find_the_difference(StringS,StringT):
    '''
        This function will take 2 string input & will provides extra character added to the second string.
    '''
    try:
        ListStringT = list(StringT)
        for i in StringS:
            if i in ListStringT:
                ListStringT.pop(ListStringT.index(i))
        LetterAddedToString = ",".join(ListStringT)
        print(f"find_the_difference({StringS}, {StringT}) ➞ {LetterAddedToString}")
    except Exception as e:
        print("Error : ",e)
        
find_the_difference("abcd", "abcde")
find_the_difference("", "y")
find_the_difference("ae", "aea")
```

```python
#Solution 2

from collections import OrderedDict

def count_datatypes(*args):
    '''
        This function will take unlimited input & will provides count of datatypes in list.
    '''
    try:
        TotalDataTypes = OrderedDict({'int':0,'str':0,'bool':0,'list':0,'tuple':0,'dict':0})
        for i in args:
            TotalDataTypes[type(i).__name__] = TotalDataTypes[type(i).__name__]+1
        print(f"count_datatypes({args}) ➞ {list(TotalDataTypes.values())}")
    except Exception as e:
        print("Error : ",e)
        
count_datatypes(1, 45, "Hi", False)   
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
#Solution 3

def fib_str(Repetition,StringList):
    '''
        This function will take 2 input(Integer,List) & provides Fibonacci String Sequence with repetition given.
    '''
    try:
        FibonacciString = StringList
        for i in range(Repetition-len(StringList)):
            FibonacciString.append(FibonacciString[i+1]+FibonacciString[i])
        print(f"fib_str({Repetition},{StringList}) ➞ {', '.join(FibonacciString)}")
    except Exception as e:
        print("Error : ",e)
        
fib_str(3, ["j", "h"])
fib_str(5, ["e", "a"])
fib_str(6, ["n", "k"])
```

```python
#Solution 4

def ones_threes_nines(Number):
    '''
        This function will take a integer input & provides ones, threes, and nines it would at least take for the number of 
        each to add up to the given integer when multiplied by one,three or nine.
    '''
    try:
        OriginalNumber = Number
        answer = {'nines':0,'threes':0,'ones':0}
        while Number > 0 and Number < 26:
                if Number >= 9:
                    Number = Number - 9
                    answer['nines'] = answer['nines'] + 1
                elif Number >= 3:
                    Number = Number - 3
                    answer['threes'] = answer['threes'] + 1
                elif Number >= 1:
                    Number = Number - 1
                    answer['ones'] = answer['ones'] + 1
        print(f"ones_threes_nines({OriginalNumber}) ➞ \"nines:{answer['nines']}, threes:{answer['threes']}, ones:{answer['ones']}\"")
    except Exception as e:
        print("Error : ",e)
        
ones_threes_nines(10) 
ones_threes_nines(15)
ones_threes_nines(22)
```

```python
#Solution 5

def fib(Number):
    '''
        This function will take integer input & provides Fibonacci Number for given number.
    '''
    try:
        f1=1
        f2=0
        for x in range(2,Number+1):
            nextsequence=f1+f2
            f2=f1
            f1=nextsequence
        print(f" fib({Number}) ➞ {f1}")
    except Exception as e:
        print("Error : ",e)
        
fib(6)
fib(1)
fib(2)
```
