```python
#Solution 1

def GetSymbolValues(listoflist):
    try:
        d = {"#":5,"O":3,"X":1,"!":-1,"!!":-3,"!!!":-5}
        Sum = 0
        for i in listoflist:
            for j in i:
                Sum = Sum+d[j]
        if Sum<0:
            return 0
        else:
            return Sum
    except Exception as e:
        print("Error : ",e)

#GetSymbolValues([["#", "!"],["!!", "X"]])
GetSymbolValues([["!!!", "O", "!"],["X", "#", "!!!"],["!!", "X", "O"]])
```

```python
#Solution 2

def GetPermutationOfItem(*args):
    try:
        Product=1
        for i in args:
            Product = Product*i
        return Product
    except Exception as e:
        print("Error : ",e)

#GetPermutationOfItem(2,3)
#GetPermutationOfItem(3,7,4)
GetPermutationOfItem(2,3,4,5)
```

```python
#Solution 4

def CheckPrimeNumberOrNot(Number):
    try:
        if ((Number-1)%6 == 0) or ((Number+1)%6 == 0):
            return True
        else:
            return False
    except Exception as e:
        print("Error : ",e)

CheckPrimeNumberOrNot(7)
CheckPrimeNumberOrNot(56963)
CheckPrimeNumberOrNot(5151512515524)
```
