```python
#Solution 1

def meme_sum(Num1,Num2):
    
    '''This function will take 2 integer input & provide sum of those numbers as shown below.
       meme_sum(26, 39) ➞ 515
       # 2+3 = 5, 6+9 = 15
       # 26 + 39 = 515
    '''
    try:
        Newstring = ""
        StringNum1 = str(Num1)
        StringNum2 = str(Num2)
        if len(StringNum1) != len(StringNum2):
            if len(StringNum1)<len(StringNum2):
                StringNum1 = "0"*(len(StringNum2)-len(StringNum1))+StringNum1
            else:
                StringNum2 = "0"*(len(StringNum1)-len(StringNum2))+StringNum2
        for i in range(len(StringNum1)):
            Newstring = Newstring+str(int(StringNum1[i])+int(StringNum2[i]))
            
        print(f'meme_sum({Num1},{Num2}) ➞ {Newstring}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
meme_sum(26, 39)
meme_sum(122, 81)
meme_sum(1222, 30277)
```

```python
#Solution 2


def is_prime(Num):
    
    '''This function will take a integer input & returns True If Number is Prime else False.
    '''
    try:
        IsPrime = False
        if Num>1:
            for i in range(2,Num):
                if Num%i == 0:
                    break
            else:
                IsPrime = True
        elif Num == 1:
            IsPrime = True
            
        return IsPrime
    
    except Exception as e:
        print('Error Occured : ',e)


def next_prime(Num):
    
    '''This function will take a integer input & the next prime. If the
       number is prime, return the number itself.
    '''
    try:
        NextPrime = 0
        for i in range(Num, (Num*2)+1):
            if is_prime(i):
                NextPrime = i
                break
        print(f'next_prime({Num}) ➞ {NextPrime}')      
    except Exception as e:
        print('Error Occured : ',e)
        
next_prime(12)
next_prime(24)
next_prime(11)
```

```python
#Solution 3

def ave_spd(UpTime,UpHillSpeed,DownHillSpeed):
    
    '''This function will take 3 integer input(Time,UphillSpeed,DownhillSpeed) & returns the average speed traveled given an uphill time,
       uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate
       as an integer (mph). No rounding is necessary.
    '''
    try:
        Distance = UpHillSpeed*(UpTime/60)
        DownTime = Distance/DownHillSpeed
        AvgSpeed = int((2*Distance)/((UpTime/60)+DownTime))
        print(f'ave_spd{UpTime,UpHillSpeed,DownHillSpeed} ➞ {AvgSpeed}')
               
    except Exception as e:
        print('Error Occured : ',e)
        
ave_spd(18, 20, 60)
ave_spd(30, 10, 30)
ave_spd(30, 8, 24)
```

```python
#Solution 4

def kempner(Num):
    
    '''This function will take a integer input & returns the smallest integer greater than zero 
        whose factorial is exactly divided by the number.
    '''
    try:
        kempner = 0
        for i in range(1,Num+1):
            Factorial = 1
            for j in range(1, i + 1):
                Factorial = Factorial*j
            if Factorial%Num == 0:
                kempner = i
                break
        
        print(f'kempner({Num}) ➞ {kempner}')
               
    except Exception as e:
        print('Error Occured : ',e)
        
kempner(6)
kempner(10)
kempner(2)
kempner(5)
```

```python
#Solution 5

def boxes(List):
    
    '''This function will take a List input & returns number of boxes to pack all of the items.
    '''
    try:
        Boxes = 0
        Evaluation = 0
        Check = 0
        for i in range(len(List)):
            Evaluation = Evaluation+List[i]
            if i == len(List)-1:
                if List[i]+List[i-1] <= 10:
                    Boxes = Boxes+1
                Check = Evaluation+List[i]
            else:
                Check = Evaluation+List[i+1]
            if Evaluation <= 10 and Check > 10:
                Evaluation = 0
                Boxes = Boxes+1
        
        print(f'boxes({List}) ➞ {Boxes}')
               
    except Exception as e:
        print('Error Occured : ',e)
        
boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2])
```
