```python
#Solution 1


def valid_rondo(String):
    
    '''This function will take a string inputs & it check string is a valid Rondo Form or not.
    '''
    try:
        NewString = String.replace('A','')
        NewString = list(NewString)
        IsValidRandoForm = False
        if len(NewString) == 1:
                IsValidRandoForm = True
        else:
            if ((String.startswith('A') and String.endswith('A')) and (len(String)>1)):
                for i in range(len(NewString)-1):
                    if ord(NewString[i]) < ord(NewString[i+1]):
                        IsValidRandoForm= True
                    else:
                        IsValidRandoForm = False
                        break
            else:
                IsValidRandoForm = False
                        
        print(f'valid_rondo({String}) ➞ {IsValidRandoForm}')
            
    except Exception as e:
        
        print('Error Occured : ',e)
        
valid_rondo("ABACADAEAFAGAHAIAJA")
valid_rondo("ABA")
valid_rondo("ABBACCA")
valid_rondo("ACAC")
valid_rondo("A")
```

```python
#Solution 2


def sentence_searcher(Sentence,Word):
    
    '''This function will take 2 string inputs(Sntence,word) & returns the whole of the first sentence which
       contains a specific word. Include the full stop at the end of the sentence.
    '''
    try:
        Result = '""'
        for i in Sentence.split(". "):
            if (Word in i.upper() or Word in i.lower()):
                Result = i if "." in i else i+"."
                break
            else:
                Result = Result
                
        print(f'sentence_searcher{Sentence,Word} ➞ {Result}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
txt = "I have a cat. I have a mat. Things are going swell."  

sentence_searcher(txt,"have")
sentence_searcher(txt,"MAT")
sentence_searcher(txt, "things")
sentence_searcher(txt, "flat")
```

```python
#Solution 3


def sum_round(Num):
    
    '''This function will take a interger input & returns the "round" of each digit
       (except if the digit is zero) as a string.
    '''
    try:
        Result = []
        for i in range(len(str(Num))):
            if str(Num)[i] !="0":
                Result.append(str(Num)[i]+len(str(Num)[i+1:])*'0') 
                
        print(f'sum_round({Num}) ➞ {" ".join(Result[::-1])}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
sum_round(101)
sum_round(1234)
sum_round(54210)
```

```python
#Solution 4


def multiplication_table(Num):
    
    '''This function will take a interger input & create N x N multiplication table, of size n provided in parameter.
    '''
    try:
        MultiplicationList = []
        for i in range(1,Num+1):
            #print(i)
            TempList = []
            for j in range(1,Num+1):
                TempList.append(j*i) 
            MultiplicationList.append(TempList)
            
        print(f'multiplication_table({Num}) ➞ {MultiplicationList}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
multiplication_table(5)
multiplication_table(1)
multiplication_table(3)
```

```python
#Solution 5


def does_rhyme(Line1,Line2):
    
    '''This function will take 2 String input & returns True if two lines rhyme and False otherwise. 
       For the purposes of this exercise,two lines rhyme if the last word from each sentence contains the same vowels.
    '''
    try:
        Vowels = 'aeiou'
        Rhyme = False
        Rhyme1 = []
        Rhyme2 = []
        for i in Line1.split(" ")[-1]:
            if i.lower() in Vowels:
                Rhyme1.append(i.lower())
                
        for j in Line2.split(" ")[-1]:
            if j.lower() in Vowels:
                Rhyme2.append(j.lower())
                
        if Rhyme1 == Rhyme2:
            Rhyme = True
            
        print(f'does_rhyme{Line1,Line2} ➞ {Rhyme}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
does_rhyme("Sam I am!", "Green eggs and ham.")
does_rhyme("Sam I am!", "Green eggs and HAM.")
does_rhyme("You are off to the races", "a splendid day.")
does_rhyme("and frequently do?", "you gotta move.")
```
