```python
#Solution 1


def transpose_matrix(List):
    
    '''This function will take a List input(List of List) & it transposes a 2D matrix.
    '''
    try:
        TransposeMatrix = []
        for i in range(len(List[0])):
            TempList = []
            for j in List:
                TempList.append(j[i])
            TransposeMatrix.append(TempList)
            
        print(f'transpose_matrix({List}) ➞ {TransposeMatrix}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
transpose_matrix([
[1, 1, 1],
[2, 2, 2],
[3, 3, 3]
])

transpose_matrix([
[5, 5],
[6, 7],
[9, 1]
])
```

```python
#Solution 2


def is_valid_hex_code(String):
    
    '''This function will take a String input & it checks whether provided string is valid hex or not.
    '''
    try:
        IsValidHexCode = False
        
        if String.startswith("#") and len(String)-1==6:
            for i in range(len(String)):
                if i != 0:
                    if (String[i] in "AaBbCcDdEeFf0123456789"):
                        IsValidHexCode = True
                    else:
                        IsValidHexCode = False
                        break
            
        print(f'is_valid_hex_code({String}) ➞ {IsValidHexCode}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
is_valid_hex_code("#CD5C5C")
is_valid_hex_code("#EAECEE")
is_valid_hex_code("#eaecee")
is_valid_hex_code("#CD5C58C")
is_valid_hex_code("#CD5C5Z")
is_valid_hex_code("#CD5C&C")
is_valid_hex_code("CD5C5C")
```

```python
#Solution 3


def mark_maths(List):
    
    '''This function will take a List input(list of math equations) & return the percentage of correct answers as a string. 
       Rounding to the nearest whole number.
    '''
    try:
        
        Result = []
        for i in List:
            i = i.split("=")
            Result.append(eval(i[0]) == int(i[1])) 
        print(f'mark_maths({List}) ➞ {str(round((sum(Result)/len(List))*100))}%')
            
    except Exception as e:
        print('Error Occured : ',e)
        
mark_maths(["2+2=4", "3+2=5", "10-3=3", "5+5=10"])
mark_maths(["1-2=-2"])
mark_maths(["2+3=5", "4+4=9", "3-1=2"])
```

```python
#Solution 4


def magic_square_game(List1,List2):
    
    '''This function will take 2 List input & it return the True & False based on following conditions: 
       Alice and Bob win the game if Alice’s numbers give an odd sum, Bob’s give an even sum, and (most important) 
       they’ve each written down the same number in the one square where their row and column intersect.
    '''
    try:
        
        Result = False
        if List2[1][List1[0]-1] == List1[1][List2[0]-1]:
            Result = True
            
        print(f'magic_square_game{List1,List2} ➞ {Result}')
            
    except Exception as e:
        print('Error Occured : ',e)
        
magic_square_game([2, "100"], [1, "101"])
magic_square_game([2, "001"], [1, "101"]) 
magic_square_game([3, "111"], [2, "011"]) 
magic_square_game([1, "010"], [3, "101"]) 
```

```python
#Solution 5

import math
def lets_meet(Distance,VelocityA,VelocityB):
    
    '''This function will take 3 integer input & it returns how much time passes until both objects meet.
    '''
    try:
        
        TotalTime = Distance/(VelocityA+VelocityB)
        Hours = math.floor(TotalTime) 
        Minutes = math.floor((TotalTime-Hours)*60)
        Seconds = math.floor(((((TotalTime)-Hours)*60)-Minutes)*60)
            
        print(f'lets_meet{Distance,VelocityA,VelocityB} ➞ {Hours}h {Minutes}min {Seconds}s')
            
    except Exception as e:
        print('Error Occured : ',e)
        
lets_meet(100, 10, 30)
lets_meet(280, 70, 80)
lets_meet(90, 75, 65)
```
