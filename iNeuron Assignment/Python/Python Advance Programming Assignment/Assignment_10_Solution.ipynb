```python
#Solution 1

def get_frame(Width,Height,Char):
    '''
        This function take 3 input(widdth,Height & Char) & returns picture frame as a 2D list.
    '''
    try:
        List2D = []
        FinalList = []
        
        if Width <= 2:
            print(f'get_frame({Width},{Height},{Char}) ➞ "Invalid"')
        else:
            for i in range(Height):
                if i == 0 or i == Height-1:
                    List2D.append([Width*Char])
                else:
                    List2D.append([Char+' '*(Width-2)+Char])
            for j in List2D:
                FinalList.append(j)
            print(f'get_frame({Width},{Height},{Char}) ➞ {FinalList}')
    except Exception as e:
        print('Error Occurred : ',e)
        
get_frame(4, 5, "#")
get_frame(10, 3, "*")
get_frame(2, 5, "0")
```

```python
#Solution 2

def boolean_and(BooleanListAnd):
    '''
        This function takes a input(List with Boolean Values) & returns True and False values, starting from
        the leftmost element and evaluating pairwise.
    '''
    try:
        BooleanListANDClone = BooleanListAnd.copy()
        for i in  range(len(BooleanListANDClone),1,-1):
            x = BooleanListANDClone.pop(0)
            y = BooleanListANDClone.pop(0)
            BooleanListANDClone.insert(0,(x and y))
        print(f'boolean_and({BooleanListAnd}) ➞ {BooleanListANDClone[0]}')
    except Exception as e:
        print('Error Occurred : ',e)
        
def boolean_or(BooleanListOR):
    '''
        This function takes a input(List with Boolean Values) & returns True and False values, starting from
        the leftmost element and evaluating pairwise.
    '''
    try:
        BooleanListORClone = BooleanListOR.copy()
        for i in  range(len(BooleanListORClone),1,-1):
            x = BooleanListORClone.pop(0)
            y = BooleanListORClone.pop(0)
            BooleanListORClone.insert(0,(x or y))
        print(f'boolean_or({BooleanListOR}) ➞ {BooleanListORClone[0]}')
    except Exception as e:
        print('Error Occurred : ',e)
        
def boolean_xor(BooleanListXOR):
    '''
        This function takes a input(List with Boolean Values) & returns True and False values, starting from
        the leftmost element and evaluating pairwise.
    '''
    try:
        BooleanListXORClone = BooleanListXOR.copy()
        for i in  range(len(BooleanListXORClone),1,-1):
            x = BooleanListXORClone.pop(0)
            y = BooleanListXORClone.pop(0)
            BooleanListXORClone.insert(0,(x ^ y))
        print(f'boolean_xor({BooleanListXOR}) ➞ {BooleanListXORClone[0]}')
    except Exception as e:
        print('Error Occurred : ',e)
        
boolean_and([True, True, False, True])
boolean_or([True, True, False, False])
boolean_xor([True, True, False, False])
```

```python
#Solution 3

def make_box(Dimension):
    '''
        This function takes a input(Dimension) & returns a box created with that dimension.
    '''
    try:
        FinalList = []
        for i in range(Dimension):
            if i == 0 or i == Dimension-1:
                FinalList.append(Dimension*'#')
                
            else:
                FinalList.append('#'+' '*(Dimension-2)+'#')
            
        print(f'make_box({Dimension}) ➞ {FinalList}')
    except Exception as e:
        print('Error Occurred : ',e)
        
make_box(5)
make_box(3)
make_box(2)
make_box(1)
```

```python
#Solution 4

def no_duplicate_letters(CommonPhrase):
    '''
        This function takes a input(common phrase) & returns False if any individual word in the phrase
        contains duplicate letters. Return True otherwise.
    '''
    try:
        IsDuplicateLetters = True
        for i in CommonPhrase.split():
            for j in i:
                if i.count(j)>1:
                    IsDuplicateLetters = False
                    break
            
        print(f'make_box({CommonPhrase}) ➞ {IsDuplicateLetters}')
    except Exception as e:
        print('Error Occurred : ',e)
        

no_duplicate_letters("Fortune favours the bold.")
no_duplicate_letters("You can lead a horse to water, but you can't make him drink.")
no_duplicate_letters("Look before you leap.")
no_duplicate_letters("An apple a day keeps the doctor away.")
```
