```python
#Solution 5
'''
    Below are the diff. between expression & statement
    
    Expression : 
                An Expression is a sequence or combination of values, variables, operators and function calls that always produces or returns a result value.
                The evaluation of an expression does not changes state and produces at least one value as a result.
        
    Statement : 
                Any instruction that a python interpreter can execute is called a Statement.
                The execution of a statement changes state.Execution of a statement may or may not produces or displays a result value, it only does whatever the statement says.
                
                Example given in question is spam = 10, here value 10 is assigning variable 'spam' is statement & its changing spam variable state but not showing result (only assigning)
'''
```
